```python
import os
from pathlib import Path
import shutil
```

```python
# Define the paths
images_path = Path("train/images")
labels_path = Path("train/labels")
labelled_path = Path("train/labelled")
```

```python
# Create the sorted folder if it doesn't exist
labelled_path.mkdir(parents=True, exist_ok=True)
```

```python
# Function to get class letter
def get_class_letter(class_num):
    return chr(class_num + ord('A'))
```

```python
# Create folders for each alphabet
for i in range(0, 26):  # A-Z
    class_folder_path = labelled_path / get_class_letter(i)
    class_folder_path.mkdir(parents=True, exist_ok=True)
```

```python
# Iterate through all label files
for label_file in labels_path.glob('*.txt'):
    # Read the class number from the label file
    with open(label_file, 'r') as file:
        first_line = file.readline().strip()
        class_num = int(first_line.split()[0])
    
    # Get the corresponding class letter
    class_letter = get_class_letter(class_num)
    
    if class_letter:
        # Create the class folder if it doesn't exist
        class_folder_path = labelled_path / class_letter
        
        # Move the image and the label file to the class folder
        image_file_name = label_file.stem + '.jpg'
        image_file_path = images_path / image_file_name
        
        if image_file_path.exists():
            shutil.move(str(image_file_path), str(class_folder_path / image_file_name))
        
        shutil.move(str(label_file), str(class_folder_path / label_file.name))
```
